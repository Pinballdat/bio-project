```python
# Install Anaconda distribution
%%shell
wget https://repo.anaconda.com/archive/Anaconda3-2024.10-1-Linux-x86_64.sh
# shasum -a 256 Anaconda3-2024.10-1-Linux-x86_64.sh
bash Anaconda3-2024.10-1-Linux-x86_64.sh -b -f -p /usr/local
```

```python
# Run fastqc to quality control raw data
%%bash
conda create -n fastqc -y
source activate fastqc
conda install -y bioconda::fastqc

mkdir Raw_fastqc
data="/content/drive/MyDrive"
fastqc $data/*.fastq.gz -o Raw_fastqc/
```

```python
# Run trimmomatic to cut adapters, bad quality positions
%%bash
conda create -n trimmomatic -y
source activate trimmomatic
conda install bioconda::trimmomatic -y

data="/content/drive/MyDrive"

mkdir Trimmomatic_paired
mkdir Trimmomatic_unpaired
mkdir Trimmomatic_template

trimmomatic PE $data/ERR7747883_forward.fastq.gz $data/ERR7747883_reverse.fastq.gz \
Trimmomatic_paired/ERR7747883_forward.paired.fastq.gz \
Trimmomatic_unpaired/ERR7747883_forward.unpaired.fastq.gz \
Trimmomatic_paired/ERR7747883_reverse.paired.fastq.gz \
Trimmomatic_unpaired/ERR7747883_reverse.unpaired.fastq.gz \
-summary Trimmomatic_template/ERR7747883.summary \
-trimlog Trimmomatic_template/ERR7747883.log \
SLIDINGWINDOW:4:20 MINLEN:100
```

```python
# Run fastqc + multiqc for paired reads
%%bash
source activate fastqc
mkdir Trimmed_fastqc
fastqc Trimmomatic_paired/* -o Trimmed_fastqc/
```

```python
# MultiQC
%%bash
conda create -n multiqc -y
source activate multiqc
conda install -y bioconda::multiqc

multiqc -o MultiQC Trimmed_fastqc/ -f
```

```python
# Run SPAdes
%%bash
wget https://github.com/ablab/spades/releases/download/v4.0.0/SPAdes-4.0.0-Linux.tar.gz
tar -xzf SPAdes-4.0.0-Linux.tar.gz
python3 /content/SPAdes-4.0.0-Linux/bin/spades.py \
-o test \
--pe1-1 Trimmomatic_paired/ERR7747883_forward.paired.fastq.gz \
--pe1-2 Trimmomatic_paired/ERR7747883_reverse.paired.fastq.gz
```

```python
# Run Quast to quality control
%%bash
conda create -n quast -y
source activate quast
conda install -y bioconda::quast

quast -o quast test/contigs.fasta
```

```python
# Remove contigs < 500 bp
%%bash
conda create -n seqkit -y
source activate seqkit
conda install -y bioconda::seqkit

seqkit seq -m 500 test/contigs.fasta > filtered_contigs.fasta
seqkit stats filtered_contigs.fasta
```

```python
# Run Busco
%%bash
conda create -y -n busco
source activate busco
conda install -c conda-forge -c bioconda busco=5.8.2 -y

busco -i filtered_contigs.fasta -m genome -o busco_results -l bacteria_odb12 -f
# conda list
```

```python
# Run prokka
%%bash
conda create -n -y prokka
source activate prokka
conda install -c conda-forge -c bioconda -c defaults prokka -y

prokka --outdir prokka --force --kingdom Bacteria --cpus 3 filtered_contigs.fasta
```

```python
# Run cogclassifier
%%bash
conda create -y -n cogclassifier
source activate cogclassifier
conda install -c conda-forge -c bioconda cogclassifier -y

protein_file="/content/prokka/PROKKA_02112025.faa"

COGclassifier -i ${protein_file} -o cogclassifier -t 2
```

```python
# Run Abricate
%%bash
conda create -y -n abricate
source activate abricate
conda install -y bioconda::abricate

contigs_file="/content/test/contigs.fasta"

abricate --db resfinder ${contigs_file} > resfinder.tab
```
